```python
from src import helpers

helper=helpers.helper()
```

```python
ticker = "AGI"

data, price_target = helper.get_daily_chart(
    ticker, startdate="2023-07-11", enddate="2024-07-11"
)

data=data.sort_values(by='date',ascending=True).reset_index(drop=True)

fig = helper.create_candlestick_graph(data)

order=2

localmin, localmax,extremas = helper.getextremes(data, order=order)

fig = helper.add_price_target(fig,data, price_target)
fig = helper.add_circles(fig, data, localmin,yparam='low',color='red')
fig = helper.add_circles(fig, data, localmax,yparam='high',color='green')
fig = helper.draw_extreme_lines(fig, data, extremas)
fig = helper.draw_support_lines(fig, data, localmax=localmax, localmin=localmin, maxage=90)

fig.update_layout(autosize=True, height=850, title_text=ticker)

fig.show()
```
